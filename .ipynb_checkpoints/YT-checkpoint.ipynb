```python
import os
import cv2
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Function to load images from folder
def load_images_from_folder(folder, target_size=(600, 1200), valid_extensions=('.png', '.jpg', '.jpeg', '.bmp')):
    images = []
    skipped_files = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if any(img_path.lower().endswith(ext) for ext in valid_extensions):
            img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)  # Load image
            if img is not None:
                img = cv2.resize(img, target_size)  # Resize image to target size
                images.append(img)
            else:
                print(f"Warning: Failed to load image '{img_path}'")
                skipped_files.append(filename)
        else:
            print(f"Skipping non-image file: '{img_path}'")
            skipped_files.append(filename)
    
    if not images:
        print(f"No images loaded from folder: '{folder}'")
    
    if skipped_files:
        print(f"Skipped {len(skipped_files)} non-image files: {skipped_files}")
    
    return np.array(images)

# Paths to the folders containing images
folder_HH = r'C:\Users\Jimi2\Implementation\Images\HH'
folder_LL = r'C:\Users\Jimi2\Implementation\Images\LL'

# Load and preprocess images
images_HH = load_images_from_folder(folder_HH)
images_LL = load_images_from_folder(folder_LL)

# Check if images were loaded successfully
if images_HH is not None and images_LL is not None:
    # Assuming labels are defined appropriately (binary classification)
    labels_HH = np.ones(len(images_HH))  # Example labels for HH images (1 for HH)
    labels_LL = np.zeros(len(images_LL))  # Example labels for LL images (0 for LL)

    # Combine images and labels
    images_combined = np.vstack((images_LL, images_HH))
    labels_combined = np.hstack((labels_LL, labels_HH))

    # Normalize images
    images_combined = images_combined.astype('float32') / 255.0

    # Print number of loaded images
    print('Number of HH images loaded:', len(images_HH))
    print('Number of LL images loaded:', len(images_LL))

    # Assuming you want to reshape to (-1, 600, 1200, 3)
    # Check the number of channels in the images
    num_channels_HH = images_HH.shape[-1] if len(images_HH.shape) == 4 else 1
    num_channels_LL = images_LL.shape[-1] if len(images_LL.shape) == 4 else 1

    # Reshape the images accordingly
    images_HH = images_HH.reshape((-1, 600, 1200, num_channels_HH))
    images_LL = images_LL.reshape((-1, 600, 1200, num_channels_LL))

    # Print shapes to verify
    print('HH images shape:', images_HH.shape)
    print('LL images shape:', images_LL.shape)

    # Define a simple CNN model
    model = Sequential([
        Conv2D(16, (3, 3), activation='relu', input_shape=(600, 1200, 3)),
        MaxPooling2D(),
        Conv2D(32, (3, 3), activation='relu'),
        MaxPooling2D(),
        Conv2D(16, (3, 3), activation='relu'),
        MaxPooling2D(),
        Flatten(),
        Dense(256, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    # Compile the model
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the model
    model.fit(images_combined, labels_combined, epochs=5, batch_size=64)

    # Evaluate the model
    loss, accuracy = model.evaluate(images_combined, labels_combined, verbose=2)
    print(f"Loss: {loss}, Accuracy: {accuracy}")

else:
    print("Error: Images could not be loaded. Check the folder paths and contents.")
```
